```python

import pylab as pl
import numpy as np
 
# values x and y give values at z
xmin = 1; xmax = 4; dx = 1
ymin = 1; ymax = 3; dy = .5
x,y = np.meshgrid(np.arange(xmin,xmax,dx),np.arange(ymin,ymax,dy))
z = x*y
 
# transform x and y to boundaries of x and y
x2,y2 = np.meshgrid(np.arange(xmin,xmax+dx,dx)-dx/2.,np.arange(ymin,ymax+dy,dy)-dy/2.)
 
# pcolormesh without x and y just uses indexing as labels
pl.subplot(121)
pl.pcolormesh(z)
pl.title("Wrong ticks")
 
# pcolormesh with x and y values gives a wrong plot, x and y are treated as boundaries
pl.subplot(122)
pl.title("Wrong: x,y as values")
pl.pcolormesh(x,y,z)
 
pl.figure()
# using the boundaries gives correct plot
pl.subplot(121)
pl.title("Right: x,y as boundaries")
pl.pcolormesh(x2,y2,z)
pl.axis([x2.min(),x2.max(),y2.min(),y2.max()])
 
# using the boundaries gives correct plot
pl.subplot(122)
pl.title("Correct ticks")
pl.pcolormesh(x2,y2,z)
pl.axis([x2.min(),x2.max(),y2.min(),y2.max()])
pl.xticks(np.arange(xmin,xmax,dx))
pl.yticks(np.arange(ymin,ymax,dy))
 
pl.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import fetch_olivetti_faces

from sklearn.linear_model import  LinearRegression,Ridge

from sklearn.neighbors import KNeighborsRegressor

#ensemble 综合算法库  随机森林 应用到复杂的数据分析上 Extra  额外的tree  
from sklearn.ensemble import  ExtraTreesRegressor
```

```python
face=fetch_olivetti_faces()
face
```

```python
x_data=face.data
y_target=face.target
```

```python
x_data.shape
```

```python
plt.figure(figsize=(2,2))
plt.imshow(x_data[0].reshape((64,64)),cmap='gray')
```

```python
from sklearn.model_selection import  train_test_split

X_train,x_test,y_train,y_test=train_test_split(x_data,y_target,train_size=0.95)
```

```python
#X_train 人脸数据，一份为二，上半部分和下半部分
#上班部分 训练数据，下半部分  是训练数据所对应的结果

face_up_train=X_train[:,:2048]
face_down_train=X_train[:,2048:]
```

```python
plt.figure(figsize=(2,2))
plt.imshow(face_train[0].reshape((32,64)),cmap='gray')
```

```python

#将x_test 一分为二
face_up_test=x_test[:,:2048]

#真是结果
face_down_test=x_test[:,2048:]
```

```python
knn=KNeighborsRegressor()
knn.fit(face_up_train,face_down_train)
```

```python
y_=knn.predict(face_up_test)
```

```python
y_.shape
```

```python
def show_predict_result(y_):
    plt.figure(figsize=(5*2,2*2))
    for i in range(5):
    #第一行绘制真实的数据
        face_true=x_test[i].reshape((64,64))
    
        axes=plt.subplot(2,5,i+1)
        axes.imshow(face_true,cmap='gray')
        plt.axis('off')
    
    #第二行绘制预测的数
    #需要将人脸上半部分的数据和预测出来下半部分的数据进行拼接
        face_up=face_up_test[i]
        face_down=y_[i]
    
        face_predict=np.hstack((face_up,face_down)).reshape((64,64))
        axes=plt.subplot(2,5,i+6)
        axes.imshow(face_predict,cmap='gray')
        plt.axis('off')
    
```

```python
show_predict_result(y_)
```

```python
#使用ridge
# knn.fit(face_up_train,face_down_train)
ridge=Ridge(alpha=1)
ridge.fit(face_up_train,face_down_train)
ridge_y_=ridge.predict(face_up_test)
```

```python
estimators={'knn':KNeighborsRegressor(),
           'linear':LinearRegression(),
           'extratree':ExtraTreesRegressor(),
           'ridge':Ridge()}
```

```python
#使用for 循环，对数据进行训练和预测
#以字典的形式来保存数据

y_=dict()

for key,estimators in estimators.items():
    estimators.fit(face_up_train,face_down_train)
    y=estimators.predict(face_up_test)
    y_[key]=y
```

```python
#将预测的结果进行可视化
#还有半张脸
cols=4+1+1
row=5
plt.figure(figsize=(cols*2,row*2))

for i in range(row):
    #输出真是的数据
    true_face=x_test[i].reshape((64,64))
    axes=plt.subplot(row,cols,i*cols+1)
    if i:
        axes.imshow(true_face,cmap='gray')
    else:
        axes.imshow(true_face,cmap='gray')
        axes.set_title('true face')
    plt.axis('off')
    
    #第二列
    face_half=face_up_test[i].reshape((32,64))
    axes=plt.subplot(row,cols,i*cols +2)
    if i:
        axes.imshow(face_half,cmap='gray')
    else:
        axes.imshow(face_half,cmap='gray')
        axes.set_title('half face')
    plt.axis('off')
    
    for j,key in enumerate(y_):
        
        predict_down_face=y_[key]
        
        down=predict_down_face[i]
        up=face_up_test[i]
        
        predict_face=np.hstack((up,down)).reshape((64,64))
        
        axes=plt.subplot(row,cols,i*cols+3+j)
        
        if i:
            axes.imshow(predict_face,cmap='gray')
        else:
            axes.imshow(predict_face,cmap='gray')
            axes.set_title(key)
        plt.axis('off')
        
        
        
```
